```python
from binance.spot import Spot
from datetime import datetime
import time
import polars as pl

def date_to_ms_timestamp(date_str):
    try:
        date_obj = datetime.strptime(date_str, "%d/%m/%y")
        timestamp_s = time.mktime(date_obj.timetuple())
        timestamp_ms = int(timestamp_s * 1000)

        return timestamp_ms
    except ValueError as e:
        print(f"Erreur: {e}. Assurez-vous que la date est au format 'jj/mm/aa'.")
    return None


client = Spot()
df = pl.DataFrame(
            data=client.klines(
                symbol="LTCBTC",
                interval="1w",
                limit=100,
                startTime=date_to_ms_timestamp("20/01/20"),
                endTime=date_to_ms_timestamp("31/08/20"),
            ),
        )
df
```
